### 6. Identify the significant variables by performing statistical analysis using p-values

Statistical significance is a determination that a relationship between two or more variables is caused by something other than chance.   
We will test the independance between 'City' and 'Complaint Type' attributes.   
H0: there is no relationship between 'City' and 'Complaint Type'.   

```python
from scipy import stats
```

```python
statistic, pvalue, dof, expected_freqn = stats.chi2_contingency(df_new)
print('Statistic:', statistic)
print('PValue:', pvalue)

if pvalue < 0.05:
    print("Reject Null Hypothesis H0 there is no relationship between 'City' and 'Complaint Type'")
else:
    print("Accept Null Hypothesis H0 there is no relationship between 'City' and 'Complaint Type'")
```

### 7. Perform a Kruskal-Wallis H test

Kruskalâ€“Wallis H test: one-way ANOVA on ranks.   
The one-way ANOVA tests the Null Hypothesis H0 that all groups have the same population mean.   
This test will be applied to the average Request Closing Time.   
Each sample is represented by a Complaint Type (10 with the most complaints).   

```python
top10_complaints
```

```python
top10_samples = [df[df['Complaint Type'] == x]['Request Closing Time'].dropna() for x in top10_complaints.index]
statistic, pvalue = stats.f_oneway(*top10_samples,)
print('Statistic:', statistic)
print('PValue:', pvalue)

if pvalue < 0.05:
    print("Reject Null Hypothesis H0 that all groups have the same population mean")
else:
    print("Accept Null Hypothesis H0 that all groups have the same population mean")
```
